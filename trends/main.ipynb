```python
import feedparser
import pandas as pd

rss_urls = [
    "https://trends.google.com/trending/rss?geo=IT&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-23&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-65&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-75&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-77&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-78&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-72&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-45&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-36&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-62&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-42&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-25&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-57&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-67&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-21&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-88&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-82&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-32&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-52&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-55&hours=24",
    "https://trends.google.com/trending/rss?geo=IT-34&hours=24"
]

dicto = {
    "IT": "IT",
    "IT-23": "Aosta",
    "IT-65": "Abruzzo",
    "IT-75": "Apulia",
    "IT-77": "Basciliata",
    "IT-78": "Calabria",
    "IT-72": "Campania",
    "IT-45": "Emilia",
    "IT-36": "Friuli",
    "IT-62": "Lazio",
    "IT-42": "Liguria",
    "IT-25": "Lombardy",
    "IT-57": "Marche",
    "IT-67": "Molise",
    "IT-21": "Piedmont",
    "IT-88": "Sardinia",
    "IT-82": "Sicily",
    "IT-32": "Tirol",
    "IT-52": "Tuscany",
    "IT-55": "Umbria",
    "IT-34": "Veneto"
}
```

```python
entries = []
for url in rss_urls:
    feed = feedparser.parse(url)
    for entry in feed.entries:
        code = url.split("=")[1].split("&")[0]
        region = dicto[code]
        entries.append({
            'Title': entry.title,
            "traffic": entry.ht_approx_traffic,
            "link": entry.ht_news_item_url,
            "published": entry.published,
            "region": region
        })
df = pd.DataFrame(entries)
```
